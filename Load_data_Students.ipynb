```python
sns.set_theme(style="whitegrid")

# Load the example planets dataset
#planets = sns.load_dataset(df_groupby)

cmap = sns.cubehelix_palette(rot=-.2, as_cmap=True)
g = sns.relplot(
    data=df_groupby,
    x=df_groupby[0].values, y=df_groupby.index,
    hue=df_groupby['ST004D01T'].values,aspect=4,
    sizes=(100, 500)
)
g.fig.suptitle('Magazines',x=0.6)
```

```python
#Which of the following statements best describes how you read books (on any topic)?
print(df_origin_schools_spain['ST168Q01HA'].value_counts())
df_readbooks = df_origin_schools_spain.groupby(['ST004D01T', 'ST167Q01IA']).size().reset_index()
print(type(df_readbooks))
#df_readbooks = df_readbooks.to_frame()
print(df_readbooks)
print(df_readbooks[0])
print(df_readbooks['ST167Q01IA'])
print(df_readbooks['ST004D01T'])
```

```python
sns.set_theme(style="whitegrid")

# Draw a nested barplot by species and sex
g = sns.catplot(
    data=df_readbooks, kind="bar",
    x=df_readbooks['ST167Q01IA'], y=df_readbooks[0], hue=df_readbooks['ST004D01T'],
    ci="sd", palette="dark", alpha=.6, height=12
)
g.despine(left=True)
g.set_axis_labels("", "How many do you read books?")
g.legend.set_title("How many do you read books?")
```
